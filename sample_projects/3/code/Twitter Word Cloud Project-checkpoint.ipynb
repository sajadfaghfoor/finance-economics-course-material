```python
# import Requests library:
import requests

# import BeautifulSoup library:
from bs4 import BeautifulSoup as BS

# import Numpy library:
from numpy import *

# import WordCloud_Fa library:
from wordcloud_fa import WordCloudFa

# import Regex library:
import re

# import Image library:
from PIL import Image

# import Hazm library:
from __future__ import unicode_literals
from hazm import *


# create Wordcloud:
class create_wordcloud:

    def EO(self):
        string_wordcloud = ''
        url =  'https://www.eghtesadonline.com/newsstudios/archive/?query=&dateRange%5Bstart%5D%5Bday%5D=01&dateRange%5Bstart%5D%5Bmonth%5D=01&dateRange%5Bstart%5D%5Byear%5D=1397&dateRange%5Bend%5D%5Bday%5D=22&dateRange%5Bend%5D%5Bmonth%5D=10&dateRange%5Bend%5D%5Byear%5D=1400&order=publish_time&chk%5B%5D=on&categories%5B%5D=7&categories%5B%5D=3&categories%5B%5D=8&categories%5B%5D=6&categories%5B%5D=41&categories%5B%5D=29&categories%5B%5D=206&categories%5B%5D=182&categories%5B%5D=205&categories%5B%5D=120&chk%5B%5D=on&types%5B%5D=1&types%5B%5D=3&types%5B%5D=4&types%5B%5D=13&types%5B%5D=14'
        page = requests.get(url)
        soup = BS(page.text, 'html.parser')
        e1 = soup.find_all('a', {'class' : 'fnb fn15 clr04'})
        for e in e1:
          e2 = e['href']
          page_news = requests.get('https://www.eghtesadonline.com' + e2)
          soup_news = BS(page_news.text, 'html.parser')
          e3 = soup_news.find_all('div', {'id' : 'news-content'})
          for div in e3:
            e4 = div.find_all('p')
            for p in e4:
              string_wordcloud += p.text
        return string_wordcloud


    def TN(self):
        string_wordcloud = ''
        page_b = requests.get('https://tejaratnews.com/category/%d8%a8%d8%a7%d8%b2%d8%a7%d8%b1/%d8%a8%d9%88%d8%b1%d8%b3?all=true')
        page_n = requests.get('https://tejaratnews.com/category/%d8%a8%d8%a7%d8%b2%d8%a7%d8%b1/%d8%a8%d9%88%d8%b1%d8%b3/%d9%86%d9%85%d8%a7%d8%af%d9%87%d8%a7')
        soup_b = BS(page_b.text, 'html.parser')
        soup_n = BS(page_n.text, 'html.parser')
        e1 = soup_b.find_all('article', {'class' : 'news-media media news-media__row news-media__l'})
        for a in e1:
          e2 = a.find_all('a')
          for h in e2:
            e3 = h['href']
            page_news = requests.get(e3)
            soup_news = BS(page_news.text, 'html.parser')
            e4 = soup_news.find_all('article', {'class' : 'single-post-main'})
            for e in e4:
              e5 = e.find_all('p')
              for p in e5:
                string_wordcloud += p.text
        e1 = soup_n.find_all('article', {'class' : 'news-media media news-media__row news-media__l'})
        for a in e1:
          e2 = a.find_all('a')
          for h in e2:
            e3 = h['href']
            page_news = requests.get(e3)
            soup_news = BS(page_news.text, 'html.parser')
            e4 = soup_news.find_all('article', {'class' : 'single-post-main'})
            for e in e4:
              e5 = e.find_all('p')
              for p in e5:
                string_wordcloud += p.text
        return string_wordcloud
    

    def BN(self):
        string_wordcloud = ''
        page = requests.get('https://www.boursenews.ir/fa/archive/section?sec_id=3')
        soup = BS(page.text, 'html.parser')
        e1 = soup.find_all('sction', {'class' : 'list-box1-cover row'})
        for e in e1:
            e2 = e.find_all('a')
            for href in e2:
              e3 = href['href']
              page_news = requests.get(e3)
              soup_news = BS(page_news.text, 'html.parser')
              e4 = soup_news.find('div', {'style' : 'direction: rtl;'})
              e5 = e4.find_all('p')
              for p in e5:
                string_wordcloud += p.text
        return string_wordcloud
        

    def DE(self):
        string_wordcloud = ''
        url = 'https://donya-e-eqtesad.com/'
        page = requests.get('https://donya-e-eqtesad.com/%D8%A8%D8%AE%D8%B4-%D8%AE%D8%A8%D8%B1-64')
        soup = BS(page.text, 'html.parser')
        e1 = soup.find_all('div', {'class' : 'service-news-list'})
        for div in e1:
          e2 = div.find('a')
          e3 = e2['href']
          page_news = requests.get(url + e3)
          soup_news = BS(page_news.text, 'html.parser')
          e4 = soup_news.find('article', {'id' : 'news-page-article'})
          e5 = e4.find_all('p')
          for p in e5:
            string_wordcloud += p.text
        return string_wordcloud


    def SE(self):
        string_wordcloud = ''
        page = requests.get('https://www.sena.ir/archive')
        soup = BS(page.text, 'html.parser')
        e1 = soup.find_all('li', {'class' : 'https://www.sena.ir/archive'})
        for a in e1:
          e2 = a.find(a)
          e3 = e2['href']
          page_news = requests.get(e3)
          soup_news = BS(page_news.text, 'html.parser')
          e4 = soup_news.find_all('article', {'class' : 'box no-header item full-img'})
          e5 = e4.find_all('p')
          for p in e5:
            string_wordcloud += p.text
        return string_wordcloud
    
    
    def remove_extra_words(self, string_wordcloud):
        normalizer = Normalizer()
        normalizer.normalize(string_wordcloud)
        stemmer = Stemmer()
        stemmer.stem(string_wordcloud)
        
        patterns = ['من', 'تو', 'او', 'ما', 'شما', 'آنها', '.', '.', ':', '،', '؛', '"', 'است', 'می باشد', 'شد', 'در', 'از', 'با', 'بر', 'برای', 'خود', 'خبرگزاری', 'اعلام', 'کرد', 'اگر', 'شاید', 'کند']
        for pattern in patterns:
            replace = ''
            new_string = re.sub(pattern, replace, string_wordcloud) 
        
    
    def crate(self, string_wordcloud):
        with open('string_wordcloud.txt', 'w') as file:
          file.write(string_wordcloud)
        wc = WordCloudFa(width=10000, height=10000, font_path='B Koodak Bold_0.ttf', persian_normalize=True)
        with open('string_wordcloud.txt', 'r') as file:
          text = file.read()
        cloud = wc.generate(text)
        image = cloud.to_image()
        image.save('persian-wordcloud.png')

        
# Run program:
string_wordcloud = ''
cw = create_wordcloud()
string_wordcloud += cw.EO()
string_wordcloud += cw.TN()
string_wordcloud += cw.BN()
string_wordcloud += cw.DE()
string_wordcloud += cw.SE()
cw.remove_extra_words(string_wordcloud)
cw.crate(string_wordcloud)
```
